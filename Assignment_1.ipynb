## Task 1: Take a paragraph of 500 words and apply tokenization, filter stop words and perform stemming operation on the paragraph. 

## Task 2: Find the frequency of the words as well.


```python
import nltk
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
paragraph = """Narendra Damodardas Modi, born 17 September 1950 is an Indian politician who is the 14th and current Prime Minister of India since 26 May 2014. Modi was the chief minister of Gujarat from 2001 to 2014 and is the Member of Parliament (MP) for Varanasi. He is a member of the Bharatiya Janata Party (BJP) and of the Rashtriya Swayamsevak Sangh (RSS), a right wing Hindu nationalist paramilitary volunteer organisation. He is the longest-serving prime minister outside the Indian National Congress. Modi was born and raised in Vadnagar in northeastern Gujarat, where he completed his secondary education. He was introduced to the RSS at the age of eight. At the age of 18, he was married to Jashodaben Modi, whom he abandoned soon after, only publicly acknowledging her four decades later when legally required to do so. Modi became a full-time worker for the RSS in Gujarat in 1971. The RSS assigned him to the BJP in 1985 and he rose through the party hierarchy, becoming general secretary in 1998. In 2001, Modi was appointed Chief Minister of Gujarat and elected to the legislative assembly soon after. His administration is considered complicit in the 2002 Gujarat riots, and has been criticised for its management of the crisis. According to official records, a little over 1,000 people were killed, three-quarters of whom were Muslim; independent sources estimated 2,000 deaths, mostly Muslim. A Special Investigation Team appointed by the Supreme Court of India in 2012 found no evidence to initiate prosecution proceedings against him. While his policies as chief minister were credited for encouraging economic growth, his administration was criticised for failing to significantly improve health, poverty and education indices in the state."""
```

```python
paragraph
```

# Tokenization

```python
nltk.download('punkt')
```

```python
sentences = nltk.sent_tokenize(paragraph)
sentences
```

```python
words = word_tokenize(paragraph.lower())
words
```

# Stemming

```python
from nltk import PorterStemmer
stemmer=PorterStemmer()
```

# Filtering Stop Words

```python
from nltk.corpus import stopwords
```

```python
words_in_paragraph = word_tokenize(paragraph)
words_in_paragraph
```

```python
# regular expression
import re
```

```python
corpus=[]
for i in range(len(sentences)):
    text=re.sub('[^a-zA-Z]', ' ', sentences[i])
    text=text.lower()
    corpus.append(text)

corpus
```

```python
stemmed_words = []

for i in corpus:
    words=nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            stemmed_words.append(stemmer.stem(word))

stemmed_words
```

# Frequency

```python
from collections import Counter
```

```python
word_frequencies = Counter(stemmed_words)
```

```python
print("Word Frequencies:", word_frequencies)
```
