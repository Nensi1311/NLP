## Using NLTK

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

text = """Technology has become an integral part of our daily lives, influencing how we communicate, work, and even think. In today’s fast-paced world, it is nearly impossible to imagine life without the advancements brought about by technological innovations. From the way we connect with friends and family to how businesses operate, technology has revolutionized every aspect of human existence. Communication, for instance, has undergone a dramatic transformation. Gone are the days when people relied solely on letters or landline phones to stay in touch. Today, we have smartphones, emails, social media platforms, and instant messaging apps that allow us to connect with anyone, anywhere, at any time. These tools have not only made communication faster but also more accessible and efficient, bridging geographical barriers that once seemed insurmountable. 
In the workplace, technology has streamlined operations, increased productivity, and opened up new opportunities. Computers, software applications, and automation tools have replaced manual processes, enabling businesses to achieve tasks that once required significant time and effort in just a fraction of the time. For example, data analysis, inventory management, and customer relationship management are now handled by advanced software solutions, allowing companies to focus on strategic growth and innovation. Moreover, remote working, which was once a rarity, has become a norm for many industries, thanks to collaboration tools like video conferencing, project management platforms, and cloud computing. This flexibility has not only improved work-life balance for employees but also expanded the talent pool for employers, as geographical location is no longer a limiting factor.
Education, too, has greatly benefited from technology. E-learning platforms, virtual classrooms, and digital resources have made education more accessible to people from all walks of life. Students can now attend classes from the comfort of their homes, access a wealth of information online, and even pursue degrees from international universities without leaving their hometowns. This democratization of knowledge has empowered individuals to upskill themselves and stay relevant in a competitive job market. Additionally, teachers and educators have access to innovative tools that enhance the learning experience, such as interactive whiteboards, virtual reality simulations, and gamified lessons.
On a personal level, technology has enriched our lives in countless ways. From wearable fitness trackers that monitor our health to smart home devices that offer convenience and security, the benefits are endless. Entertainment options have also exploded, with streaming services, online gaming, and virtual reality experiences providing immersive ways to unwind and relax. However, despite its many advantages, the rapid proliferation of technology has also raised concerns about its impact on privacy, mental health, and social interactions. Excessive screen time, cyberbullying, and data breaches are just a few of the challenges that need to be addressed to ensure that technology continues to serve as a force for good.
In conclusion, technology is a double-edged sword that has the power to transform our lives for better or worse. While it has undoubtedly brought immense benefits, it is crucial to use it responsibly and mindfully. By striking a balance and addressing its challenges, we can harness the full potential of technology to build a brighter and more connected future for everyone.."""

text
```

```python
sentences = nltk.sent_tokenize(text)
sentences
```

```python
for sentence in sentences:
    words = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(words)
    print(pos_tags)
```

## Using spacy

```python
import spacy
nlp = spacy.load('en_core_web_sm')

doc = nlp(text)
for token in doc:
    print(token.text, token.pos_)
```

# Sentiment Analysis

```python
from textblob import TextBlob

for sentence in sentences:
    blob = TextBlob(sentence)
    sentiment = blob.sentiment
    print(f"Sentence: {sentence}\nSentiment: {sentiment}\n")
```

# Sentence Division and Similarity Scores

```python
documents = [sentence for sentence in sentences[:5]]
```

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from gensim import corpora, similarities
from gensim.utils import simple_preprocess
```

```python
processed_docs = [simple_preprocess(doc) for doc in documents]
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

# Create similarity matrix
similarity_index = similarities.MatrixSimilarity(corpus)

# Print similarity scores
for i, doc in enumerate(corpus):
    sims = similarity_index[doc]
    print(f"Similarity scores for Document {i+1}: {list(enumerate(sims))}")
```
